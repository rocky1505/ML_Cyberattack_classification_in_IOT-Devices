```python
import pandas as pd
import pickle


# Load the pre-trained model
with open('/kaggle/input/afasfafaff/XG BOOST_34classes.pkl', 'rb') as file:
    model = pickle.load(file)

# Load the CSV file
csv_file_path = '/kaggle/input/fafsafasfas/predict.xlsx'
data = pd.read_excel(csv_file_path)
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer(output_distribution='uniform')
data[X_columns] = scaler.fit_transform(data[X_columns])
# Preprocess the data if needed (e.g., handle missing values, encode categorical variables)
# Ensure that the data matches the format expected by the model (e.g., same input features, scaling)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(original_labels)
predictions = model.predict(data)
predictions = label_encoder.inverse_transform(predictions)
  # Assuming the model.predict method exists for your model

# Optionally, postprocess the predictions if needed

# Output the predictions (e.g., print, save to a file)
print(predictions)
```

```python
from tensorflow.keras.models import load_model
from keras.models import model_from_json
model_path = '/kaggle/input/vsdvsvds/CNN34classfinal'
def load_model(model_path):
    with open(model_path + ".json", "r") as json_file:
        loaded_model_json = json_file.read()
    loaded_model = model_from_json(loaded_model_json)
    # Load weights into the model
    loaded_model.load_weights(model_path + ".weights.h5")
    return loaded_model
# Load the model
model = load_model(model_path)
csv_file_path = '/kaggle/input/fafsafasfas/predict.xlsx'
data = pd.read_excel(csv_file_path)
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer(output_distribution='uniform')
data[X_columns] = scaler.fit_transform(data[X_columns])

# Preprocess the data if needed (e.g., handle missing values, encode categorical variables)
# Ensure that the data matches the format expected by the model (e.g., same input features, scaling)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(original_labels)
predictions = model.predict(data)
predictions = predictions.argmax(axis=1)
predictions = label_encoder.inverse_transform(predictions)
  # Assuming the model.predict method exists for your model

# Optionally, postprocess the predictions if needed

# Output the predictions (e.g., print, save to a file)
print(predictions)
```
